Given a string S, find the longest palindromic substring in S.
You may assume that the maximum length of S is 1000,
and there exists one unique longest palindromic substring.

```python
def isPalindrome(s):
    lenS = len(s)
    middle = int(lenS / 2)
    for i in range(middle, 0, -1):
        start = 0 + i - 1
        end = lenS - i
        if s[start] != s[end]:
            return False

    return True

def longestPalindrome(s):
    """
    :type s: str
    :rtype: str
    """

    lenS = len(s)

    if lenS == 1:
        return s

    lenS2 = lenS - 1
    longestStart = 0
    longestEnd = 0
    longestLength = 0
    for position in range(0, lenS2):
        isPotential = False
        start = position
        end = position + 1
        if s[start] == s[end]:
            while True:
                # for the same character
                endT = end + 1
                if endT >= lenS or s[start] != s[endT]:
                    break

                end = endT

            isPotential = True

        # skip middle for example aabaa
        endT = end + 1
        if endT < lenS and s[start] == s[endT]:
            end = endT
            isPotential = True

        while isPotential:
            startT = start - 1
            endT = end + 1

            if startT < 0 or endT >= lenS or s[startT] != s[endT]:
                break

            start = startT
            end = endT

        if isPotential:
            length = end - start + 1

            if length > longestLength:
                longestLength = length
                longestStart = start
                longestEnd = end + 1

    return s[longestStart:longestEnd]
```
